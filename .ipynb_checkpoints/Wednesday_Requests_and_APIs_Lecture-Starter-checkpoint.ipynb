```python
base_url = "http://api.weatherapi.com/v1"
api_method = "/current.json"
api_key = "1ac9901266fa4ab48f9185441220510"
city = 'Chicago'


api_url = f"{base_url}{api_method}?key={api_key}&q={city}"

print(api_url)
```

```python
res = requests.get(api_url)
weather_data = res.json()
weather_data
```

```python
current_temp = weather_data['current']['temp_f']
print(current_temp)

feels_like = weather_data['current']['feelslike_f']
print(feels_like)

current_condition = weather_data['current']['condition']['text']
print(current_condition)
```

```python
class CityWeather:
    def __init__(self, name, region, current, feels_like, condition):
        self.city = name
        self.region = region
        self.current_temp = current
        self.feels_like = feels_like
        self.condition = condition
        
    def __repr__(self):
        return f"<CityWeather|{self.city}>"
    
    def __str__(self):
        return f"It is currently {self.condition} and {self.current_temp} in {self.city}, {self.region}. \
It feels like {self.feels_like}"
```

```python
class WeatherAPI:
    base_url = "http://api.weatherapi.com/v1"
    
    def __init__(self, api_key):
        self.api_key = api_key
        
    def __get(self, city, api_method, **params): # Private method - can only be called from other methods within the class
        request_url = f"{self.base_url}{api_method}?key={self.api_key}&q={city}"
        if params:
            for key, value in params.items():
                request_url += f"&{key}={value}"
        weather_response = requests.get(request_url)
        if weather_response.ok:
            return weather_response.json()
        else:
            print('There was an error!')
    
    def get_current_weather(self, city):
        weather_data = self.__get(city, '/current.json')
        if weather_data:
            city_name = weather_data['location']['name']
            region = weather_data['location']['region']
            current_temp = weather_data['current']['temp_f']
            feels_like = weather_data['current']['feelslike_f']
            current_condition = weather_data['current']['condition']['text']
            city_weather = CityWeather(city_name, region, current_temp, feels_like, current_condition)
            return city_weather
    def get_forecast_weather(self, city, days=1):
        weather_data = self.__get(city, '/forecast.json', days=days)
        return weather_data
        
client = WeatherAPI('a06e65a8a9904be182e203809230103')
client.get_current_weather('chicago')      
```

#### Pokemon API Wrapper

Create a python wrapper for the Pokemon API. It should take in a pokemon name and display the pokemon with its height and weight

```python
pokemon_res = requests.get('https://pokeapi.co/api/v2/pokemon/charizard')

pokemon_res
```

```python
pokemon_res.json()
```

```python
class Pokemon:
    def __init__(self, name):
        self.name = name
        
        pokemon_res = requests.get(f"https://pokeapi.co/api/v2/pokemon/{name}")
        pokemon = pokemon_res.json()
        
        self.height = pokemon["height"]
        self.weight = pokemon["weight"]
        
    def __str__(self):
        return f"the height is {self.height} feet and weighing in at {self.weight} pounds, the pokemon {self.name.title()} was selected!"
```

```python
charizard = Pokemon ('charizard')
```

```python
print(charizard)
```

```python
pikachu = Pokemon('pikachu')
```

```python
print(pikachu)
```

```python
class PokeAPI:
    base_url = "http://pokeapi.co/api/v2/"
    
    def _get(self, endpoint, id_or_name):
        request_url = f"{self.base_url}{endpoint}{id_or_name}"
        res = requests.get(request_url)
        if res.ok:
            return res.json()
        else:
            print('There was an error')
            
    def get_poke_info(self, poke_name):
        data = self._get('pokemon/', str(poke_name).lower())
        height = data.get('height')
        weight = data.get('weight')
        pokemon = Pokemon(poke_name, height, weight)
        return pokemon


class Pokemon:
    def __init__(self,name,height,weight):
        self.name = name.title()
        self.height = height
        self.weight = weight
        
    def __str__(self):
        return self.name
    
    def __repr__(self):
        return f"<Pokemon| {self.name}>"
        
poke_client = PokeAPI()
charizard = poke_client.get_poke_info('pikachu')
```
